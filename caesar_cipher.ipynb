# Libraries

# Function

```python
def caesar():
    print(logo)
    decode_again = True
    
    alphabet = []
    for i in string.ascii_letters:
        alphabet += i
    alphabet = alphabet * 2
    
    
    while decode_again == True:
        cipher_direction = input("Type 'encode' to encrypt, type 'decode' to decrypt:\n").lower()
        start_text = input("Type your message:\n")
        shift_amount = int(input("Type the shift amount: "))
        
        if shift_amount > 26:
            shift_amount = shift_amount % 26
        
        end_text = ""
        if cipher_direction == "decode":
            shift_amount *= -1
        
        for char in start_text:
            if char in alphabet:
                position = alphabet.index(char)
                new_position = position + shift_amount
                end_text += alphabet[new_position]
                
            else: 
                end_text += char
                
        print(f"Here is the {cipher_direction}d result: {end_text}")
        caesar_continue = input("Type 'yes' if you want to continue. Otherwise, type 'no'.\n")
        if caesar_continue == "yes":
            decode_again = True
        else:
            decode_again = False
```
